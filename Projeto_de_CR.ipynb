Para ver o quão conectadas estão as espécies podemos analisar  a densidade do grafo. A densidade mostra o quão conectada é uma rede. A densidade deste grafo é baixa porque cada espécie se liga apenas com suas espécies descendentes  e como o grafo é uma árvore filogenética não tem muitas conexões entre espécies
A densidade pode ser observada na figura 2

```python
densidade = nx.density(g)
print("A densidade é:",densidade)
```

A densidade de um grafo explica o nível geral de conectividade da rede. No caso, observamos que o grafo possui densidade igual a 0,25, ou seja, indica que um quarto de todas as ligações possíveis entre os vértices foram feitas, no entanto é um número relativamente baixo, porém esperado para uma árvore filogenética.

###Figura 3

A modularidade é uma medida que avalia a conectividade de partições de um grafo segundo o estudo de sua configuração (mesma sequência de grau de nós). Em resumo, quanto maior a modularidade de uma partição, melhor sua tendência de agrupamento.

Em nosso grafo, constatamos 7 grupos de vértices que variam seu peso de 0 a 6, obtivemos modularidade de 0.778, muito próximo de 1, o que significa a maior conectividade possível entre as partições.

###Figura 4

Centralidade de proximidade está relacionada com o tempo que uma informação leva para ser compartilhada por todos os vértices na rede, e a intermediação de um vértice pode ser considerada como o controle da comunicação entre todos os demais pares de vértices da rede, no caso, a informação seria a informação genética passada pelos ancestrais. É dada pelo inverso da soma das distâncias do vértice (espécie) escolhido a todos os demais vértices do grafo.

Observamos 4 espécies com conectividade de proximidade igual a 1 (o máximo) e 17 espécies igual a 0 (o mínimo), as demais, encontram-se distribuidas entre esses extremos.

Harmonic Closeness Centrality Distribution.png

###Figura 5

Centralidade de intermediação de uma espécie no grafo, é dado ao somar todas as intermediações parciais do vertice no grafo. O valor encontrado será definido como a centralidade de intermediação. Indica o potencial que um ponto da rede, no caso a espécie, tem para controlar o fluxo de informações genéticas entre os pares de espécies da árvore.

Podemos observar se estender de 0 até mais de 1400 entre as espécies, havendo praticamente 1 ou 2 espécies compartilhando centralidade, com exceção de 27, que nao possuem centralidade de proximidade.
