#Leitura de csv

```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/superstore_data.csv'
dados_mercado = pd.read_csv(url)
dados_mercado.head()
```

#Mudar o nome das colunas

```python
dados_mercado = dados_mercado.rename(columns = {'Year_Birth' : 'Data de Nascimento'})
dados_mercado = dados_mercado.rename(columns = {'Education' : 'Grau de Escolaridade'})
dados_mercado = dados_mercado.rename(columns={'Marital_Status':'Estado civil'})
dados_mercado
```

#Mudar os dados das linhas de uma coluna


```python
dados_mercado['Grau de escolaridade'] = dados_mercado['Grau de escolaridade'].replace('Graduation', 'Graduação')
dados_mercado
```

#Ler arquivos em SQL

```python
import sqlalchemy
```

```python
from sqlalchemy import create_engine, MetaData, Table, inspect
```

```python
sqlalchemy.__version__
```

Usar o banco de dados local

```python
pip install --upgrade 'sqlalchemy<2.0'
```

```python
engine = create_engine('sqlite:///:memory:')
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv'
```

```python
dados = pd.read_csv(url)
```

```python
dados.head()
```

Coloca os dados dentro do banco de dados

clientes é o nome da tabela que sera criada

```python
dados.to_sql('clientes',engine,index=False)
```

```python
inspector = inspect(engine)
```

```python
print(inspector.get_table_names())
```

```python
query = 'SELECT * FROM clientes WHERE Grau_escolaridade = "Ensino superior" '
```

Ler uma consulta

```python
empregados = pd.read_sql(query,engine)
```

```python
empregados
```

Salvar a consulta como uma tabela

```python
empregados.to_sql('empregados2',con=engine,index=False)
```

Ler uma tabela

```python
pd.read_sql_table('empregados2',engine)
```

```python
pd.read_sql_table('empregados2',engine, columns=['ID_Cliente','Grau_escolaridade','Rendimento_anual'])
```

```python
query = 'SELECT * FROM clientes'
```

```python
pd.read_sql(query,engine)
```

Comando que pede para executar o comando

```python
query = 'DELETE FROM clientes WHERE ID_CLIENTE=5008804'
with engine.connect() as conn:
  conn.execute(query)
```
